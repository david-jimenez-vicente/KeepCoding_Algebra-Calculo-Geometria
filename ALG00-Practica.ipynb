**Problema**: Implementar las operaciones de los multiconjuntos (utilizando las librer铆as y estructuras de datos vistas en el curso).

**Datos:**

Un multiconjunto es un conjunto en el que un elemento puede repetirse, es decir, cada elemento posee una multiplicidad (un n煤mero natural) que indica cu谩ntas veces el elemento es miembro del conjunto. Por ejemplo, en el multiconjunto `{a, a, b, b, b, c}`, las multiplicidades de los miembros a, b, y c son 2, 3, y 1, respectivamente.

Al igual que los conjuntos, poseen las siguientes caracter铆sticas y operaciones:
- Cardinalidad: indica el n煤mero de elementos del multiconjunto. Por ejemplo, la cardinalidad del multiconjunto `{a, a, b, b, b, c}` es 6 (la suma de sus multiplicidades).
- Inserci贸n: permite insertar una ocurrencia de un elemento en el multiconjunto.
- Eliminaci贸n: permite eliminar una ocurrencia de un elemento del multiconjunto.
- Comparaci贸n: compara dos multiconjuntos para determinar si son iguales.
- Pertenencia: determina si un elemento pertenece al multiconjunto.
- Subconjunto: determina si un multiconjunto es subconjunto de otro.
- Uni贸n: conjunci贸n de todos los elementos de dos multiconjuntos (sumando sus multiplicidades si un elementos est谩 en los dos).
- Intersecci贸n: elementos que est谩n en los dos multiconjuntos qued谩ndonos con la multiplicidad m谩s peque帽a.
- Diferencia: restar a un multiconjunto los elementos de otro.

```python
### TODO: Crear una funci贸n que dada una lista devuelva un multiconjunto
### El multiconjunto que devuelve puede crearse con la estructura de datos que se quiera (incluso una lista)
### TU RESPUESTA ABAJO

def multiconjunto(elementos):
    """
    Devuelve un multiconjunto con la lista de elementos dada
    
    Argumentos:
        elementos -- lista de elementos
        
    Ejemplo:
        elementos = [1,1,1,3,3,1,4,5,1,5]
    """
    # Usar茅 una lista como estructura de datos.
    # Creo una lista vac铆a para guardar la entrada de elementos.
    multic_list = []
    # Itero por los elementos de entrada para a帽adirlos uno a uno en la nueva lista.
    for j in elementos:
        multic_list.append(j)
    # Devuelvo el multiconjunto.
    return sorted(multic_list)

mc = multiconjunto([1,1,1,3,3,1,4,5,1,5])
mc
#for key, value in mc.items():
    #print(value)
    #print(key, len([item for item in value if item]))
```

```python
### TODO: Crear una funci贸n que dado un multiconjunto y un elemento devuelva el multiconjunto con el elemento insertado
### TU RESPUESTA ABAJO

def inserta(multiconjunto, elemento):
    """
    Devuelve el multiconjunto habiendo insertado el elemento dado
    
    Argumentos:
        multiconjunto -- multiconjunto devuelto por la funci贸n creada anteriormente
        elemento -- elemento a insertar
    """
    # Por el funcionamiento de Python, hay que crear un clon de los argumentos de entrada para manipularlos.
    # Si intentamos manipular directamente la colecci贸n de entrada dar谩 un objeto al transformarlo en un type None.
    new1 = multiconjunto[:]
    # Ahora simplemente se a帽ade al final el argumento elemento
    new1.append(elemento)
    return sorted(new1)

mc = inserta(mc, 9)
mc
```

```python
### TODO: Crear una funci贸n que dado un multiconjunto y un elemento devuelva si el elemento pertenece al multiconjunto
### TODO: Crear una funci贸n que dados dos multiconjuntos devuelva si el primero es subconjunto del segundo
### TODO: Crear una funci贸n que dados dos multiconjuntos devuelva si son iguales
### TU RESPUESTA ABAJO

def pertenece(multiconjunto, elemento):
    """
    Devuelve si el elemento pertenece al multiconjunto
    
    Argumentos:
        multiconjunto -- multiconjunto devuelto por la funci贸n creada anteriormente
        elemento -- elemento a determinar si pertenece
    """
    new4 = multiconjunto[:]
    # Sencilla comprobaci贸n de existencia en una lista.
    if elemento in new4:
        return (f"El elemento {elemento} s铆 pertenece al multiconjunto {new4}")
    else:
        return (f"El elemento {elemento} no pertenece al multiconjunto {new4}")
        
def subconjunto(multiconjunto1, multiconjunto2):
    """
    Devuelve si multiconjunto1 es subconjunto de multiconjunto2
    
    Argumentos:
        multiconjunto1 -- multiconjunto devuelto por la funci贸n creada anteriormente
        multiconjunto2 -- multiconjunto devuelto por la funci贸n creada anteriormente
    """
    
    multi1 = multiconjunto1[:]
    multi2 = multiconjunto2[:]
    # Para empezar, si el primer multiset tiene mayor cardinalidad que el segundo, no puede ser subconjunto de 茅ste 煤ltimo.
    if (len(multi1) > len(multi2)):
        return (f"El multiconjunto {multi1} no es subconjunto de {multi2}")
        
    # Ahora contaremos los elementos en cada multiset, y si coincide el conteo del primero con el segundo, es un subconjunto.
    # Lo covertimos en set para comparar menos veces los elementos repetidos.
    subset1 = set(multi1)
    # Contamos cada elemento del primero y lo comparamos con el conteo en el segundo:
    for i in subset1:
        # Si no coincide, dar谩 False
        if (multi1.count(i)) > multi2.count(i):
            return (f"El multiconjunto {multi1} no es subconjunto de {multi2}")
    return (f"El multiconjunto {multi1} si es subconjunto de {multi2}")
    
def iguales(multiconjunto1, multiconjunto2):
    """
    Devuelve si multiconjunto1 es igual a multiconjunto2
    
    Argumentos:
        multiconjunto1 -- multiconjunto devuelto por la funci贸n creada anteriormente
        multiconjunto2 -- multiconjunto devuelto por la funci贸n creada anteriormente
    """
    multi3 = multiconjunto1[:]
    multi4 = multiconjunto2[:]
    # Si las listas son iguales, dar谩 resultado positivo. Si no, negativo.
    if multi3 == multi4:
        return ("Los multiconjuntos son iguales")
    return ("Los multiconjuntos no son iguales")

# Bater铆a de pruebas
print(pertenece(mc, 1))
print(pertenece(mc, 10))
print(subconjunto(multiconjunto([1,3,3,3,4]), mc))
print(subconjunto(multiconjunto([1,3,3,4]), mc))
print(subconjunto(mc, multiconjunto([1,3,3,3,4])))
print(iguales(multiconjunto([1,3,4]), mc))
print(iguales(multiconjunto([1, 1, 1, 1, 1, 3, 3, 4, 5, 5]), mc))
```

```python
### TODO: Crear una funci贸n que dados dos multiconjuntos devuelva su uni贸n
### TODO: Crear una funci贸n que dados dos multiconjuntos devuelva su intersecci贸n
### TODO: Crear una funci贸n que dados dos multiconjuntos devuelva su diferencia
### TU RESPUESTA ABAJO

def union(multiconjunto1, multiconjunto2):
    """
    Devuelve la uni贸n de los multiconjuntos
    
    Argumentos:
        multiconjunto1 -- multiconjunto devuelto por la funci贸n creada anteriormente
        multiconjunto2 -- multiconjunto devuelto por la funci贸n creada anteriormente
    """
    multi5 = multiconjunto1[:]
    multi6 = multiconjunto2[:]
    print(f"La uni贸n de {multi5} y {multi6} es:")
    return multi5 + multi6

def interseccion(multiconjunto1, multiconjunto2):
    """
    Devuelve la intersecci贸n de los multiconjuntos
    
    Argumentos:
        multiconjunto1 -- multiconjunto devuelto por la funci贸n creada anteriormente
        multiconjunto2 -- multiconjunto devuelto por la funci贸n creada anteriormente
    """
    multi7 = multiconjunto1[:]
    multi8 = multiconjunto2[:]
    # Recogemos cada elemento de un multiset si el elemento est谩 en el otro multiset. Los comprueba de uno en uno:
    inter = [i for i in multi7 if i in multi8]
    # Si hay cero coincidencias, es que es disjunto:
    if len(inter) == 0:
        return (f"{multi7} y {multi8} son multiconjuntos disjuntos")
    # Si no es disjunto, devuelve la intersecci贸n.
    print(f"La intersecci贸n de {multi7} y {multi8} es:")
    return inter
            

def diferencia(multiconjunto1, multiconjunto2):
    """
    Devuelve la diferencia de los multiconjuntos
    
    Argumentos:
        multiconjunto1 -- multiconjunto devuelto por la funci贸n creada anteriormente
        multiconjunto2 -- multiconjunto devuelto por la funci贸n creada anteriormente
    """
    multi9 = multiconjunto1[:]
    multi10 = multiconjunto2[:]
    # Creo una copia del primero para irle eliminando los elementos del segundo uno a uno:
    dife = multi9[:]
    for i in multi10:
        if i in multi9:
            dife.remove(i)
    print(f"La diferencia de {multi9} y {multi10} es:")
    return dife


print(union(multiconjunto(["a","b","c"]), mc), "\n")
print(union(multiconjunto([1,3,4]), mc), "\n")
print(interseccion(multiconjunto([1,3,3,4,30]), mc), "\n")
print(interseccion(multiconjunto([11,32,41,]), mc), "\n")
print(diferencia(mc, multiconjunto([1,2,3,4])), "\n")
print(diferencia(mc, multiconjunto([1,1,1,1,1,2,3,4])), "\n")
```

Deberas crear tu propia funci贸n para calcular el error de reconstrucci贸n, que viene definido por:

$$SSE =  \sum_{n}^{i=1}  \begin{Vmatrix}x_{i} -  \widehat{x}_i \end{Vmatrix} ^2 $$

Donde:

- $x_i$ son los valores de la matriz original X
- $\widehat{x}_i$ son los valores de la matriz reconstruida

```python
### TODO: Funci贸n para calcular el error de reconstrucci贸n
### TU RESPUESTA ABAJO

import numpy as np

def sse_score(X, X_hat):
    """
    Funci贸n para calcular el error de reconstrucci贸n
    
    Argumentos:
        X -- Matriz Original
        X_hat -- Matriz Reconstruida
        
    Ejemplo:
        X = np.array([[1, 2], [3, 4]])
        X_hat = np.array([[1.01, 1.75], [2.81, 3.99]])
        sse = sse_score(X, X_hat) # -> 0.09879
    """
    # Seguimos la f贸rmula:
    sse = np.sum((X - X_hat)**2)
    return sse
```

```python
# Prueba
X = np.array([[1, 2], [3, 4]])
X_hat = np.array([[1.01, 1.75], [2.81, 3.99]])
sse = sse_score(X, X_hat)
sse
```

Una vez que ya tenemos la funci贸n `sse` hecha, podemos pasar a construir la funci贸n que ejecutar谩 `SVD`.

```python
### TODO: Funci贸n para ejecutar SVD
### Tiene como entrada una matriz X
### Devuelve U, s, Vt

### Hint: S debe ser una matriz diagonal
### TU RESPUESTA ABAJO

def svm(X):
    """
    Funci贸n que ejecuta SVD y devuelve U, S, Vt
    
    Argumentos:
        X -- Matriz Original
        
    Ejemplo:
        X = np.array([[1, 2], [3, 4]])
        U, S, Vt = svm(X)  
        
        # -> U = array([[-0.40455358, -0.9145143 ],
        #               [-0.9145143 ,  0.40455358]])
        # -> S = array([[5.4649857 , 0.        ],
        #               [0.        , 0.36596619]])   
        # -> Vt = array([[-0.57604844, -0.81741556],
        #                [ 0.81741556, -0.57604844]])          
    """ 
    # Descomponemos la matriz de entrada usando la funci贸n svd() de numpy
    U, s, Vt = np.linalg.svd(X, full_matrices=False)
    # Diagonalizamos s
    S = np.diag(s)
    
    return U, S, Vt
```

Como hemos visto en clase, las matrices obtenidas a partir de `SVD` nos sirven para reconstruir la matriz original `X`. Para ello, construye una funci贸n que permita reconstruir la matriz original `X` a partir de `U, s, Vt`.

```python
### TODO: Funci贸n para reconstruir la matriz original a partir de U, s, Vt
### Tiene como entrada U, s, Vt
### Devuelve X_hat
### TU RESPUESTA ABAJO


def reconstruction(U, S, Vt):
    """
    Funci贸n que reconstruye la matriz original a partir de U, s, Vt
    
    Argumentos:
        U -- Matriz de Singular Vectors
        s -- Matriz de Eigenvalues
        Vt -- Matriz de Singular Vectors
        
    Ejemplo:
        U = np.array([[-0.40455358, -0.9145143 ],
                      [-0.9145143 ,  0.40455358]])
        S = np.array([[5.4649857 , 0.        ],
                      [0.        , 0.36596619]])
        Vt = np.array([[-0.57604844, -0.81741556],
                       [ 0.81741556, -0.57604844]])
        X_hat = reconstruction(U, S, Vt)
        
        # X_hat -> array([[0.99999999, 1.99999998],
        #                 [3.00000003, 4.00000001]])
    """
    # Seguimos la f贸rmula para recomponer la matriz original
    X_hat = U.dot(S.dot(Vt))
    return X_hat
```

```python
### TODO: Funci贸n que recibe una imagen A y devuelve la imagen comprimida
### Tiene como entrada A y el n煤mero de componentes para realizar la reducci贸n de dimensionalidad
### Devuelve la imagen comprimidad, el error de reconstrucci贸n y el ratio de compresi贸n

### Hint: Usa las funciones anteriormente construidas
### TU RESPUESTA ABAJO


def image_compression(A, n_components):
    """
    Funci贸n para comprimir una imagen A
    
    Argumentos:
        A -- Imagen original
        n_components -- N煤mero de componentes
        
    Ejemplo:
        A_hat, sse, comp_ratio = image_compression(A, n_components=50)
    """
    # Uso la funci贸n creada al efecto para descomponer la matriz de la imagen
    U, S, Vt = svm(A)
    
    # Segmentamos las matrices seg煤n el n煤mero de componentes escogido y las guardamos
    U_redu = U[:, :n_components]
    S_redu = S[:n_components, :n_components]
    Vt_redu = Vt[:n_components, :]

    # Reconstru铆mos la matriz de la imagen usando las matrices reducidas
    A_hat_reconst = reconstruction(U_redu, S_redu, Vt_redu)

    # Calculamos el error
    sse = sse_score(A, A_hat_reconst)

    # Calculamos el ratio de compresi贸n, que es el n煤mero de elementos de la matriz comprimida entre el n煤mero de elementos de la matriz sin comprimir
    comp_ratio = (A.shape[1]*n_components + n_components + n_components*A.shape[0]) / (A.shape[1] * A.shape[0])

    # Imprimimos la imagen original y luego la comprimida.
    plt.imshow(A, cmap=plt.cm.gray)
    print(plt.show())
    plt.imshow(A_hat_reconst, cmap=plt.cm.gray)
    print(plt.show())

    return f"El error SSE es de {sse}, y el ratio de compresi贸n es {round(comp_ratio, 2)}"
```

#### Subset del dataframe por valores

El siguiente paso es construir una funci贸n que recibe un `data_frame`, el nombre de una columna, un valor m铆nimo y un valor m谩ximo `cutoffs`. Nos devuelve un dataframe excluyendo las filas donde el valor de la columna indica est谩 fuera de los valores m铆nimos y m谩ximos que le hemos indicado.

```python
### TODO: Funci贸n para crear un nuevo subset en base a valores
### Como argumento recibe un dataframe y una lista de tuples
### Tuples: (column_name, min_value, max_value)
### Devuelve un dataframe que excluye las filas donde los valores, en la columna que le hemos indicado, exceden los valores
### que le hemos indicado
### No eliminar la fila si los valores son iguales al min/max valor
### TU RESPUESTA ABAJO

def column_cutoff(data_frame, cutoffs):
    """Crea un nuevo dataframe en base a unos l铆mites
    
    Argumentos:
        data_frame -- Dataframe Object
        cutoffs -- Lista de tuples con el siguiente formato:
        (column_name, min_value, max_value)
        
    Ejemplo:
        data_frame = read_into_data_frame('train.csv')
        # Remove data points with SalePrice < $50,000
        # Remove data points with GrLiveAre > 4,000 square feet
        cutoffs = [('SalePrice', 50000, 1e10), ('GrLivArea', 0, 4000)]
        selected_data = column_cutoff(data_frame, cutoffs)
    """
    # Hago una copia del dataframe original
    df = pd.DataFrame(data_frame)
    # Para cada tupla en la lista de entrada de los cutoffs....
    for i, j in enumerate(cutoffs):
        # ....extraemos los 3 valores....
        column_name, min_value, max_value = j
        #.... y en cada loop guardo el propio dataframe ya filtrado por la primera tupla, que ser谩 de nuevo reducido
        # en el siguiente loop al filtrarse con la siguiente tupla.
        df = df[df[column_name].between(min_value, max_value)]
        
    return df
```

#### M铆nimos Cuadrados / Least Squares

Ahora, implementar谩s la ecuaci贸n $w_{LS}$:

<center>$w_{LS} = (X^T X)^{1}X^T y,$</center>

```python
### TODO: Funci贸n para resolver la ecuaci贸n wLS
### Toma como argumentos dos matrices, una para X y otra para y
### Asumimos que las matrices tienen las dimensiones correctas

### Paso 1: Asegurate que n > d. 
### Es decir, que el n煤mero de observaciones es mayor que el n煤mero de dimensiones.
### O lo que es lo mismo, que el n煤mero de filas de cada matriz sea mayor que el n煤mero de columnas
### Si no es as铆, debes transponer las matrices

### Paso 2: Debes a帽adir a la matriz X un vector columna del tama帽o (n x 1)

### Paso 3: Usa la ecuaci贸n de arriba para obtener wLS

### TU RESPUESTA ABAJO


def least_squares_weights(input_x, target_y):
    """Resuelve la ecuaci贸n para wLS
    
    Argumentos:
        input_x -- Matriz con los datos de entrenamiento
        target_y -- Vector con los datos de salida
        
    Ejemplo:
        import numpy as np
        training_y = np.array([[208500, 181500, 223500, 
                                140000, 250000, 143000, 
                                307000, 200000, 129900, 
                                118000]])
        training_x = np.array([[1710, 1262, 1786, 
                                1717, 2198, 1362, 
                                1694, 2090, 1774, 
                                1077], 
                               [2003, 1976, 2001, 
                                1915, 2000, 1993, 
                                2004, 1973, 1931, 
                                1939]])
        weights = least_squares_weights(training_x, training_y)
        
        print(weights)  #--> np.array([[-2.29223802e+06],
                        #              [ 5.92536529e+01],
                        #              [ 1.20780450e+03]])
                           
        print(weights[1][0]) #--> 59.25365290008861
    
    Asumimos:
        -- target_y es un vector con el mismo n煤mero de observaciones que input_x
    """
          
    # Comprobar orientaci贸n de las dimensiones de las matrices de entrada, para elegir la configuraci贸n correcta de n > d:
    if input_x.shape[0] < input_x.shape[1]:
        input_x = np.transpose(input_x)
    if target_y.shape[0] < target_y.shape[1]:
        target_y = np.transpose(target_y)

    # Creamos una matriz vertical de unos para a帽adirla a la matriz X...
    unos = np.ones((input_x.shape[0],1))
    # ...y la juntamos con X
    X = np.hstack((unos, input_x))
    
    # C谩lculo secuencial de la ecuaci贸n matricial:
    
    Xt = np.transpose(X) # Traspuesta de X
    XtX = Xt.dot(X) # Producto dot de X traspuesta por X
    XtX_inv = inverse_of_matrix(XtX) # Inversa del producto anterior
    XtX_inv_Xt = XtX_inv.dot(Xt) # Ahora lo anterior lo multiplicamos por la matriz traspuesta
    XtX_inv_Xt_y = XtX_inv_Xt.dot(target_y) # Y finalmente todo lo multiplicamos por la matriz de los resultados reales

    return XtX_inv_Xt_y
```

## 4. Linear Regression - Gradient Descent

En este ejercicio resolveras el mismo problema anterior pero usando **Gradient Descent**

**Objetivos**:
- Asegurar los fundamentos matem谩ticos detr谩s del Gradient Descent.

**Problema**: Usando datos sobre el precio de la vivienda, intentaremos predecir el precio de una casa en base a la superficie habitable con un modelo de regresi贸n.

**Datos:** [Kaggle's House Prices Dataset](https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data)

**Repaso:**

$$ RSS(w) = \sum_{n=1}^{N}[y_n-f(x_n)]^2 =  \sum_{n=1}^{N}[y_n- (w_0 + \sum_{d=1}^{D}w_dx_{nd}) ]^2 .$$

Loss function:

$$ RSS(w) = \frac{1}{2}\sum_{n=1}^{N}[y_n-f(x_n)]^2$$

Y lo que queremos es minimizar esta distancia, para que el modelo se acerque lo m谩ximo posible a los valores verdaderos.

$$\nabla RSS(w) = X^T(Xw^t-y)$$

En resumen, el gradient descendiente para una regresi贸n lineal, se basa en resolver esta ecuaci贸n de forma iterativa:

$$w^{t+1} = w^t - \eta * \nabla RSS(w)$$

#### Leer Datos

```python
import pandas as pd
import numpy as np

# Leer datos
data = pd.read_csv('./data/train.csv')

# Extraer dichas columnas
newData = data[['GrLivArea','SalePrice']]
print(newData.head())

# Contruir x - y
x = newData['GrLivArea']
y = newData['SalePrice']

# Standarizar los datos
x = (x - x.mean()) / x.std()
x = np.c_[np.ones(x.shape[0]), x] 

print("Shape of X: ", x.shape)
print("Shape of y:", y.shape)
```

#### Gradient Descent

```python
### TODO: Funci贸n para encontrar los valores w usando Gradient Descent
### Toma como argumentos: X, y, w, n_iterations, eta
### Completa la funci贸n a帽adiendo la loss funci贸n y la updating rule
### TU RESPUESTA ABAJO

def gradient_descent(x, y, w, iterations, eta):
    """Gradient descent
    
    Argumentos:
        x -- Matriz con los datos de entrenamiento
        y -- Vector con los datos de salida
        w -- Vector aleatoriamente inicializado
        iterations -- N煤mero de iteraciones
        eta -- Learning Rate
        
    Ejemplo:
        import numpy as np

        # Learning rate
        eta = 0.01 

        # N煤mero de iteraciones
        iterations = 2000 #No. of iterations

        # Seed para inicializar w
        np.random.seed(123)
        w0 = np.random.rand(2)
        
        training_y = np.array([208500, 181500, 223500, 
                                140000, 250000])
        training_x = np.array([[ 1.        ,  0.37020659],
                               [ 1.        , -0.48234664],
                               [ 1.        ,  0.51483616],
                               [ 1.        ,  0.38352774],
                               [ 1.        ,  1.29888065]])
                            
        weights, loss = gradient_descent(training_x, training_y, w0, iterations, eta)
        
        print(weights[-1])  #--> np.array([183845.82320222  40415.66453324])
    """

    # Defino dos listas para guardar los resultados hist贸ricos de los errores y de los pesos.
    # Guardar los pesos puede ser guardado dentro de listas o dentro de arrays de numpy.
    # Comparados los dos m茅todos, el guardado en lista es mucho m谩s r谩pido que el proceso en arrays.
    rss_reg = []
    w_log = []
    #w_log = np.empty((iterations, w.size))
    
    # Ahora iteramos la cantidad especificada:
    for i in range(iterations):
        
        # Predicciones para el c谩lculo del error:
        # Primero calculamos los datos predichos por los pesos que entran
        y_hat = np.dot(x, w)
        # Calculo el error sobre los datos reales
        error = y_hat - y
        # Calculo la funci贸n de p茅rdida con rss
        rss = (np.sum(error**2))/2
        # Guardamos en resultado del error en el hist贸rico
        rss_reg.append(rss)
        
        # Ahora calculamos el GD.
        # La nueva matriz de w sustituye a la anterior como entrada para el c谩lculo del error rss....
        w = w - eta * (x.T.dot(x.dot(w) - y))
        # ... y se guarda en el hist贸rico de pesos
        #w_log[i] = w
        w_log.append(w)
    
    return w_log, rss_reg
    
```

```python
import numpy as np

# Learning rate
eta = 0.01 

# N煤mero de iteraciones
iterations = 2000 #No. of iterations

# Seed para inicializar w
np.random.seed(123)
w0 = np.random.rand(2)
        
training_y = np.array([208500, 181500, 223500, 140000, 250000])
training_x = np.array([ [ 1. ,  0.37020659],
                        [ 1. , -0.48234664],
                        [ 1. ,  0.51483616],
                        [ 1. ,  0.38352774],
                        [ 1. ,  1.29888065]])
                            
weights, loss = gradient_descent(training_x, training_y, w0, iterations, eta)
        
print(weights[-1])  #--> np.array([183845.82320222  40415.66453324])
```

El siguiente paso es crear un conjunto con las palabras de ambas p谩ginas web (uni贸n), por ejemplo:

- Los perros son maravillosos...
- Los maravillosos a帽os 80...

Por tanto, el conjunto para estas dos frases ser铆a `{"los", "perros", "son", "maravillosos", "a帽os", "80"}`. Debemos realizar esto para todas las combinaciones posibles, es decir:

```python
### TODO: Escribe una funci贸n que recibe una lista de N p谩ginas web
### Y calcula todas las combinaciones posibles entre ellas, no importa el orden
### [web_1, web_2, web_3, ...]
### Devuelve una lista de tuples con las combinaciones [(web_1, web_2), (web_1, web_3), ...]

# HINT: Puedes implementar esta funci贸n como quieras pero la librer铆a itertools 
#       proporciona una funci贸n llamada `combinations` para realizar esta tarea.

### TU RESPUESTA ABAJO
import itertools
    
def combinations(url_ls):
    """Calcula todas las combinaciones posibles entre los elementos de una lista
    
    Argumentos:
        url_ls -- Lista de urls
        
    Ejemplo:
        url_ls = ['https://es.wikipedia.org/wiki/Canis_lupus_familiaris', 
        'https://es.wikipedia.org/wiki/Canis_lupus',
        'https://es.wikipedia.org/wiki/Felis_silvestris_catus']
        
        permutation = combinations(url_ls)  
    
    Requerimientos:
        Puedes implementar esta funci贸n como quieras pero la librer铆a itertools 
        proporciona una funci贸n llamada `combinations` para realizar esta tarea.
    """
    # Creamos una lista con las combinaciones de dos en dos de la entrada de la funci贸n
    combis = list(itertools.combinations(url_ls, 2))
    return combis
    
```

```python
### TODO: Escribe una funci贸n que recibe una lista con tuples
### [({'que', 'hola'}, {'que', 'es', 'guay'}), ({'que', 'hola'}, {'madrid', 'la', 'es'})]
### Y devuelve una lista con la union de los conjuntos
### [({'que', 'hola', 'es', 'guay'}), ({'que', 'hola', 'madrid', 'la', 'es'})]
### TU RESPUESTA ABAJO

def union(comb_ls):
    """Calcula la uni贸n por cada tuple de una lista 
    
    Argumentos:
        comb_ls -- Lista de tuples
        
    Ejemplo:
        comb_ls = [({'que', 'hola'}, {'que', 'es', 'guay'}), ({'que', 'hola'}, {'madrid', 'la', 'es'})]
        
        union_ls = union(comb_ls)  # -> [{'es', 'que', 'hola', 'guay'}, {'es', 'que', 'hola', 'madrid', 'la'}]
    """
    # Este m茅todo lo he pensado por si la lista de tuplas tiene un n煤mero variable de tuplas en cada ejecuci贸n.
    # Creo una lista de listas. Cada lista interior guardar谩 los conjuntos de una tupla de sets.
    lista = []
    # Esta lista contendr谩 las uniones finales de cada tupla de sets
    conjs = []
    # Para cada tupla de la lista de entrada:
    for i,j in enumerate(comb_ls):
        # Almaceno temporalmente en un solo espacio los sets de una sola tupla
        temp = []
        for conj in j:
            temp.append(conj)
        # Ahora almaceno en una sola lista interior de la lista final los sets
        lista.append(temp)
        
    # Para cada lista interna a帽ado la uni贸n de todos sus elementos a la lista de uniones
    for s in lista:
        conjs.append(set.union(*s))
        
    return(conjs)
```

Una vez que tenemos una lista de conjuntos por cada par de p谩ginas web, podemos convertir el texto de la p谩gina web a un vector.

```python
def set2vector(tokens_web1, tokens_web_2, set_web1, set_web2):
    """
    Funci贸n para convertir un conjunto a vector
    
    Argumentos:
        tokens_web1 -- Contenido tokenizado de p谩gina web 1
        tokens_web_2 -- Contenido tokenizado de p谩gina web 2
        set_web1 -- Conjunto de palabras de la p谩gina web 1
        set_web2 -- Conjunto de palabras de la p谩gina web 2
        
    Ejemplo:
        tokens_web1 = ["hola", "que", "tal", "soy", "Alex"]
        tokens_web_2 = ["hola", "me", "llamo"]
        set_web1 = {"hola", "que", "tal"}
        set_web2 = {"hola", "me", "llamo"}
        union_ls = set2vector(tokens_web1, tokens_web_2, set_web1, set_web2)  
        
    Requerimientos:
        Depende de la funci贸n `union()`
    """
    
    # Unimos los conjuntos
    join_set = union([(set_web1, set_web2)])[0]
    
    web1_array = []
    web2_array = [] 

    for word in join_set:
        if word in tokens_web1:
            web1_array.append(1)
        else:
            web1_array.append(0)
        if word in tokens_web_2:
            web2_array.append(1)
        else:
            web2_array.append(0)

    return web1_array, web2_array

tokens_web1 = ["hola", "que", "tal", "soy", "Alex"]
tokens_web_2 = ["hola", "me", "llamo"]
set_web1 = {"hola", "que", "tal", "soy", "Alex"}
set_web2 = {"hola", "me", "llamo"}
web1_array, web2_array = set2vector(tokens_web1, tokens_web_2, set_web1, set_web2)  
```

### Cosine Similarity

Por 煤ltimo, ya podemos implementar la ecuaci贸n: $\cos \theta = \frac{\overrightarrow{u} \cdot \overrightarrow{v}}{|\overrightarrow{u}||\overrightarrow{v}|}$

```python
### TODO: Escribe una funci贸n que recibe dos vectores, u y v
### Y devuelva la similaridad entre ambos vectores
###
### Paso 1: Si u y v son listas -> Convertirlo a arrays
###
### Paso 2: Calcula la similaridad entre ambos vectores
### TU RESPUESTA ABAJO

import numpy as np

def cosine_similarity(u, v):
    """Calcula la similaridad entre dos vectores
    
    Argumentos:
        u -- Vector 1
        v -- Vector 2
        
    Ejemplo:
        u = np.array([1, 2, 3])
        v = np.array([3, 2, 1])
        
        similarity = cosine_similarity(u, v)  # -> 0.71428
    """
    # Convertimos a arrays las entradas
    u = np.array(u)
    v = np.array(v)
    # Devolvemos el c谩lculo de la f贸rmula    
    return np.dot(u, v)/(np.linalg.norm(u)*np.linalg.norm(v))
```
